```python
# Elias Bernabé Turchiello
# 12/02/2023

################################################################################
# IMPORTS
################################################################################

import datetime
import pandas               as pd
import numpy                as np
import matplotlib.pyplot    as plt






################################################################################
# FUNCTIONS
################################################################################

def show_header(message, wdt, chr):
    tm = datetime.datetime.now().time()
    print(f'{chr}' * wdt)
    print(f'{chr} {message.upper()} {tm}')
    print(f'{chr}' * wdt)


def spliter():
    print(f'-' * 120)
    print(f'-' * 120)


def get_db_xlsx(file):
    return pd.read_excel(file)


def get_db_csv(file, dlmt=';'):
    return pd.read_csv(file , delimiter=dlmt)





################################################################################
# RUN
################################################################################

# SHOW A HEADER
show_header('extra - loading data ', 120, '#')
start           = datetime.datetime.now().time()
print(f'start          -> loading... - {start}'.upper())

covid19      = get_db_csv('/content/total_deaths.csv', ',')
finish       = datetime.datetime.now().time()
print(f'covid19     -> success!! - {finish}'.upper())
```

```python
# GET BRASIL DATA
show_header('Brasil data', 120, '#')
result = covid19.loc[covid19['Brazil'] >= 1, ['date', 'Brazil']]
display(result)


show_header('Brasil first case of covid 19', 120, '#')
# GET BRASIL FIRST CASE
result = covid19.loc[covid19['Brazil'] >= 1, ['date', 'Brazil']].head(1)
display(result)




# SHOW A BRASIL CHART
show_header('Brasil a chart of covid 19 number of case increase', 120, '#')
plt.figure(figsize=(15,5))
plt.plot(covid19['Brazil']) # Data
plt.title('covid 19'.upper()) # Title of chart
plt.xlabel('Number of days') # X label
plt.ylabel('Number os case') # Y label
plt.legend('COVID') # Set a legend
plt.axhline(covid19['Brazil'].mean(), color='orange', linestyle='dashed', linewidth=1.5) # Show a horizontal line
plt.grid() # Enable the grid
plt.show() # Ploat the chart
```
