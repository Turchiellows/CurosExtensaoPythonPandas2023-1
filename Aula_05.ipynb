```python
# Elias BernabÃ© Turchiello
# 10/02/2023

################################################################################
# IMPORTS
################################################################################

import datetime
import pandas               as pd
import numpy                as np
import matplotlib.pyplot    as plt
import seaborn              as sns
from openpyxl               import load_workbook
from openpyxl.styles        import PatternFill, Font, Color, Alignment





################################################################################
# FUNCTIONS
################################################################################

def show_header(message, wdt, chr):
    tm = datetime.datetime.now().time()
    print(f'{chr}' * wdt)
    print(f'{chr} {message.upper()} {tm}')
    print(f'{chr}' * wdt)


def spliter():
    print(f'-' * 120)
    print(f'-' * 120)


def get_db_xlsx(file):
    return pd.read_excel(file)


def get_db_csv(file, dlmt=';'):
    return pd.read_csv(file , delimiter=dlmt)





################################################################################
# RUN
################################################################################

# SHOW A HEADER
show_header('aula 04 - loading data ', 120, '#')
start           = datetime.datetime.now().time()
print(f'start          -> loading... - {start}'.upper())

# LOAD THE DATABASES FROM SOME FILES
fuel            = get_db_xlsx('/content/ca202102_20230207120945.xlsx')
finish          = datetime.datetime.now().time()
print(f'fuel           -> success!! - {finish}'.upper())


# population      = get_db_csv('/content/ibge_num_habitantes_estimado.csv', ';')
# finish          = datetime.datetime.now().time()
# print(f'population     -> success!! - {finish}'.upper())
```

```python
from os import execl
################################################################################
# CHARTS
################################################################################

chart=False
if chart:
    # CREATE A SIMPLE BAR CHART
    plt.hist(fuel['Valor de Venda']) # Data
    plt.title('Prece of fuels') # Title of chart
    plt.xlabel('Value of fuel') # X label
    plt.ylabel('Quantity of collects') # Y label
    plt.legend('Fuel price') # Set a legend
    plt.autoscale(enable=True) # Enable auto scale
    plt.axvline(fuel['Valor de Venda'].mean(), color='orange', linestyle='dashed', linewidth=1.5) # Show a a vertical line
    plt.grid() # Enable the grid
    sns.despine() # Format the chart
    plt.show() # Ploat the chart




    # CREATE A NEW CHART
    fuel_group = fuel['Valor de Venda'].groupby(by=fuel['Produto']).mean() # Data
    plt.figure(figsize=(10,5)) # Set size of chart
    fuel_group.plot(
        title='Average by fuel',
        legend='Average',
        alpha=0.6,
        kind="barh",
        xlabel="Fuel type",
        ylabel="Price",
        color="red"
    ) # Chart configuration
    plt.grid() # Enable the grid
    sns.despine() # Format the chart
    plt.show() # Ploat the chart

   







################################################################################
# EXCEL
################################################################################

# CREATE A SMAL TABLE
file    = '/content/Sumary.xlsx'
plan    = fuel['Valor de Venda'].groupby(by=fuel['Produto']).mean()
plan.to_excel(file, 'Sum')

file    = '/content/Sumary.xlsx'
wb      = load_workbook(file) # Get the fie
ws      = wb['Sum'] # Select the sheet
ccc     = PatternFill('solid', fgColor='cccccc') # Deffine a collor
aaa     = PatternFill('solid', fgColor='15F264') # Deffine a collor

# COLOR COLUMNS
for x in range(1, 8):
    x   = f'A{x}'
    ws[x].fill = ccc

# COLOR ROWS
for y in range(65, 67):
    Z   = f'{chr(y).upper()}1'
    #print(Z)
    ws[Z].fill = aaa


# COLOR TEXT BY A RULER
max_row     = ws.max_row
num_linha   = 2
while (num_linha <= max_row):
    cord = 'B' + str(num_linha)
    if ws[cord].value >= 6.24:
        ws[cord].font = Font(bold=True, color='FF0000')

    # Stop
    num_linha += 1

# SAVE INTO THE SAME FILE
wb.save(file)

print('Done!')
```
